```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("inshorts data preprocessed.csv")
```

```python
df.head()
```

```python
df['length of headline'] = [len(i) for i in df['Headline_lemma']]
```

```python
df['length of short'] = [len(i) for i in df['Short_lemma']]
```

```python
df.head()
```

```python
df['Subject'].unique()
```

```python
# Distribution of the word count for Politics subject
sns.distplot(df[df['Subject']=='Politics']['length of headline'], kde=False)
```
